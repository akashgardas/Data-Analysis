```python
import pandas as pd
```

```python
file_path = '/content/drive/MyDrive/Tekworks/Online_Retail.csv'
df = pd.read_csv(file_path, encoding='latin1')
df
```

### Descriptive Analysis

```python
df.info()
```

```python
df.shape
```

### Handling Missing Data

```python
df.isnull().sum()
```

```python
# Drop: Missing 'CustomerID'
df.dropna(subset=['CustomerID'], inplace=True)
df.isnull().sum()
```

```python
df.shape
```

### Data Parsing / Format Correction

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.info()
```

### Incorrect Data / Outlier Data fix

```python
# Drop: Neg 'Quantity'
df = df[df['Quantity'] > 0]
df.shape
```

```python
df.describe()
```

```python
# Drop: Zero 'UnitPrice'
df = df[df['UnitPrice'] > 0]
df.shape
```

### Handling Duplicates

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates().copy()
df.duplicated().sum()
```

### Filtering Data

```python
df.describe()
```

```python
# Clip: 'Quantity'
df['Quantity'] = df['Quantity'].clip(0, 1000)
df.describe()
```

```python
# Clip: 'UnitPrice'
df['UnitPrice'] = df['UnitPrice'].clip(0, 100)
df.describe()
```

### Sorting data

```python
df['Revenue'] = df['UnitPrice'] * df['Quantity']
df.describe()
```

```python
df.sort_values(by=['Revenue'], inplace=True, ascending=False)
df.head()
```

### Group & Aggregation

```python
# Total revenues by 'Country'
df.groupby('Country')['Revenue'].sum().sort_values(ascending=False).head()
```

```python
# Mean revenue per order
df['Revenue'].mean()
```

```python
import numpy as np
price_bins = [0, 2, 5, 20, np.inf]
price_labels = ['Cheap', 'Medium', 'Expensive', 'Very Expensive']

# Create a new column 'PriceRange' with the price buckets
df['PriceRange'] = pd.cut(df['UnitPrice'], bins=price_bins, labels=price_labels)
df
```

### Correlation / Relationships

```python
df[['Quantity', 'UnitPrice', 'Revenue']].corr()
```

### Rename & String Operations

```python
df.rename(columns={'InvoiceNo': 'Invoice_No', 'StockCode': 'Product_ID', 'UnitPrice': 'Price'}, inplace=True)
df.columns
```

```python
df['Description'] = df['Description'].str.lower()
df.head()
```

### Desciptive Stats

```python
df.describe()
```

```python
df['Country'].value_counts()
```

### Advanced Filtering / Querying

```python
df.query('Revenue > 4000 and Country == "United Kingdom"')
```

```python
df['Quantity'].between(10, 100).sum()
```
