# Matplotlib

- **Histogram**: Displaying the distribution of data
- **Line Plot**: Trend over time
- **Bar Chart**: Comparing quantities across different categories
- **Scatter Plot**: Relationship between two variables
- **Pie Chat**: Proportions of a whole

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = "../Data Source/car_dataset.csv"
df = pd.read_csv(path)
df
```

## Histogram
- Distribution of a continuous variable

```python
plt.figure(figsize=(6, 5))
plt.hist(df['Price (INR)'], bins=10, color='green')
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# histogram for engine sizes
plt.figure(figsize=(10, 4))
plt.hist(df['Engine_Size (L)'], bins=df['Engine_Size (L)'].unique().sort(), color='orange', edgecolor='black')
plt.title('Distribution of Engine Sizes in Inventory')
plt.xlabel('Engine Size (Liters)')
plt.ylabel('Number of cars')
plt.show()
```

```python
df['Engine_Size (L)'].unique()
```

## Line Plot
- Visualizing continuous data points over a continuous range.

```python
# filter data for Toyota cars
df_Toyota = df[df['Make'] == 'Toyota']
df_Toyota
```

```python
# group by year and calculate avg price 
avg_price_toyota = df_Toyota.groupby('Year')['Price (INR)'].mean()
avg_price_toyota
```

```python
# line plot
plt.plot(avg_price_toyota.index, avg_price_toyota, marker='o', color='blue', linestyle='-', linewidth=2)
plt.title('Avg price of toyota cars')
plt.xlabel('Year')
plt.ylabel('Avg Price (in INR)')
plt.show()
```

```python
# group by year and calculate avg price
avg_price_year = df.groupby('Year')['Price (INR)'].mean()
avg_price_year
```

```python
plt.plot(avg_price_year.index, avg_price_year, marker='o', color='red', linestyle='-', linewidth=2)
plt.title('Avg car price over years')
plt.xlabel('Year')
plt.ylabel('Avg Price (in INR)')
plt.show()
```

```python
# avg mileage over years
avg_mileage = df.groupby('Year')['Mileage (in miles)'].mean()
avg_mileage
```

```python
# line plot
plt.plot(avg_mileage.index, avg_mileage, marker='o', linewidth=3)
plt.title('Avg Mileage Over Years')
plt.xlabel('Year')
plt.ylabel('Avg Mileage')
plt.show()
```

## Bar Chart
- To show size of different categories.
- Easy comparison between categories.

```python
# avg price by make
avg_price_make = df.groupby('Make')['Price (INR)'].mean()
avg_price_make
```

```python
# bar plot
plt.figure(figsize=(10, 4))
plt.bar(avg_price_make.index, avg_price_make, color='skyblue', edgecolor='black')
plt.title('Avg Car Prices by Make')
plt.xlabel('Make')
plt.ylabel('Avg Price')
plt.show()
```

## Scatter Plot
- To compare two numerical variables

```python
plt.scatter(df['Mileage (in miles)'], df['Price (INR)'], color='green', marker='o')
plt.title('Mileage vs Price')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

```python
# car price vs year
plt.scatter(df['Year'], df['Price (INR)'])
```

```python
# engine size vs car price
plt.scatter(df['Engine_Size (L)'], df['Price (INR)'])
```

## Pie Chart
- To represent percentage or proportion of each category

```python
# number of cars for each make
make_counts = df['Make'].value_counts()
make_counts
```

```python
plt.pie(make_counts, labels=make_counts.index, autopct='%1.1f%%', colors=plt.cm.Paired(range(len(make_counts))))
plt.title('Distribution of Car Makes')
plt.show()
```

```python
# number of cars for each engine size
engine_size_counts = df['Engine_Size (L)'].value_counts()
engine_size_counts
```

```python
plt.pie(engine_size_counts, labels=engine_size_counts.index, autopct='%1.1f%%', colors=plt.cm.Paired(range(len(engine_size_counts))))
plt.title('Distribution of Engine Size')
plt.show()
```

## Customizing Plots

```python
# avg price by make
avg_price_make = df.groupby('Make')['Price (INR)'].mean()
avg_price_make
```

```python
# top 3 car makes
top_makes = avg_price_make.nlargest(3)
```

```python
# Plotting bar with customized colors
colors = ['red' if make in top_makes.index else 'grey' for make in avg_price_make.index]

plt.figure(figsize=(10, 4))
plt.bar(avg_price_make.index, avg_price_make, color=colors, edgecolor='black')
plt.show()
```

```python
# Define marker styles for different makes
markers = {
    'Toyota': 'o',
    'Honda': 's',
    'Ford': '^',
    'BMW': 'D',
    'Audi': 'X',
    'Mercedes': '*',
    'Volkswagen': '+',
    'Subaru': 'H',
    'Nissan': 'p',
    'Chevrolet': 'v'
}

# Plotting the scatter plot
for make in df['Make'].unique():
    subset = df[df['Make'] == make]
    plt.scatter(
        subset['Mileage (in miles)'],
        subset['Price (INR)'],
        marker=markers.get(make, 'o'),
        label=make
    )

plt.xlabel('Mileage (in miles)')
plt.ylabel('Price (in INR)')
plt.title('Price vs. Mileage by Car Make')
plt.legend()
plt.show()
```
