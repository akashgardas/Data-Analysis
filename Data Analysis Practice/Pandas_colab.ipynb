## Data Merging & Joining

### Merging
- This involves combining DataFrames by aligning them on a common column(s) or index, similar to how SQL JOINs work. You can perform different types of merges, such as inner, outer, left, and right joins, depending on how you want to combine the data.

image.png

### Joining
This is another method in Pandas that specifically combines DataFrames based on their indices.

```python
# DataFrame 1: Engagement Metrics
df_engagement = pd.DataFrame({
    'PostID': [1, 2, 3, 4],
    'Likes': [1500, 2000, 1200, 1600],
    'Comments': [200, 300, 180, 220],
    'Shares': [80, 120, 70, 90]
})

# DataFrame 2: Post Details
df_details = pd.DataFrame({
    'PostID': [1, 2, 3, 4],
    'Post_Type': ['Image', 'Video', 'Text', 'Image'],
    'Post_Date': ['2024-08-01', '2024-08-02', '2024-08-03', '2024-08-04']
})

# Merging the DataFrames on 'PostID'
merged_df = pd.merge(df_engagement, df_details, on='PostID')

print("Merged DataFrame:")
merged_df
```

```python
# Performing a left join
left_join_df = pd.merge(df_engagement, df_details, on='PostID', how='left')

print('Left Join DataFrame:')
left_join_df
```

```python
# Performing an outer join
outer_join_df = pd.merge(df_engagement, df_details, on='PostID', how='outer')

print('Outer Join DataFrame:')
outer_join_df
```

```python
# setting index for both DataFrames
df_engagement_indexed = df_engagement.set_index('PostID')
df_details_indexed = df_details.set_index('PostID')

# Joining on indices
joined_df = df_engagement_indexed.join(df_details_indexed)
print('Joined DataFrame on Indices:')
joined_df
```
