# Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Data Loading and Inspection

```python
df = pd.read_csv('drive/MyDrive/Tek works/patient_health_data.csv')
```

## 2. Descriptive Statistics

```python
df.describe()
```

## 3. Univariate Analysis

```python
# Age
sns.set_style('whitegrid')
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'])

plt.subplot(1, 2, 2)
sns.boxplot(df['Age'])

plt.show()

print(f"Skewness: {df['Age'].skew().round(3)}\nKurtosis: {df['Age'].kurtosis().round(3)}")
```

```python
# BMI
sns.set_style('whitegrid')
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['BMI'])

plt.subplot(1, 2, 2)
sns.boxplot(df['BMI'])

plt.show()
print(f"Skewness: {df['BMI'].skew().round(3)}\nKurtosis: {df['BMI'].kurtosis().round(3)}")
```

### Insights:
- Seems like outliers are present.

```python
# HealthScore
sns.set_style('whitegrid')
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['HealthScore'])

plt.subplot(1, 2, 2)
sns.boxplot(df['HealthScore'])

plt.show()
print(f"Skewness: {df['HealthScore'].skew().round(3)}\nKurtosis: {df['HealthScore'].kurtosis().round(3)}")
```

```python
# Smoking
sns.set_style('whitegrid')
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Smoking')

plt.subplot(1, 2, 2)
sns.countplot(data=df, x='Smoking', hue='Gender')
plt.show()
```

```python
# ExcerciseFrequency
sns.set_style('whitegrid')
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='ExerciseFrequency')

plt.subplot(1, 2, 2)
sns.countplot(data=df, x='ExerciseFrequency', hue='Gender')
plt.show()
```

```python
# MedicalConditions
sns.set_style('whitegrid')
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='MedicalConditions')

plt.subplot(1, 2, 2)
sns.countplot(data=df, x='MedicalConditions', hue='Gender')
plt.show()
```
