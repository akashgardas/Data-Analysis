## 1A. Numerical Columns
(Item_price, connected_handling_time, CSAT Score)

```python
num_df = df[['Item_price', 'connected_handling_time', 'CSAT Score']]
num_df.describe()
```

```python
num_df.info()
```

### 1. Show me the distribution of Item_price.
 Histogram

```python
# distribution of item_price
sns.histplot(num_df['Item_price'], kde=False)
plt.title('Distribution of Item Price')
plt.show()
```

### 2. How does CSAT Score vary overall?
 Boxplot

```python
sns.boxplot(y=num_df['CSAT Score'])
plt.title('CSAT Score Distribution')
plt.show()
```

### 3. What is the spread of connected_handling_time values?
 Histogram + boxplot

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.histplot(num_df['connected_handling_time'])
plt.title('Distribution of connected_handling_time')

plt.subplot(1, 2, 2)
sns.boxplot(y=num_df['connected_handling_time'])
plt.title('connected_handling_time Distribution')
plt.show()
```

### 4. Can I see a pie chart of CSAT scores by their value?
  Pie chart

```python
value_counts = num_df['CSAT Score'].value_counts()
value_counts
```

```python
plt.pie(x=value_counts, labels=value_counts.index, autopct='%1.1f%%')
plt.title('CSAT Score Distribution')
plt.show()
```

### 5. What are the most common price ranges for items?
 Binned bar chart

```python
bins = [num_df['Item_price'].min(), 5000, 20000, num_df['Item_price'].max()]
labels = ['Low', 'Med', 'High']

num_df['Price_Range'] = pd.cut(df['Item_price'], bins=bins, labels=labels)
num_df.groupby('Price_Range')['Price_Range'].count()
```

```python
sns.barplot(data=num_df, x='Price_Range', y='Item_price')
plt.title('Price Range Distribution')
plt.show()
```

## 3. Temporal Columns

### 21. How many tickets were created per day?
 Line plot

```python
tickets_per_day = temporal_df.groupby('Survey_response_Date')['Survey_response_Date'].count()
tickets_per_day.head()
```

```python
plt.figure(figsize=(12, 4))
sns.lineplot(x=tickets_per_day.index, y=tickets_per_day)
plt.title('Tickets Created per Day')
plt.xticks(rotation=90)
plt.show()
```

### 22. Show daily trend of average CSAT Score.
 Line plot (rolling average optional)

```python
daily_csat_avg = df.groupby('Survey_response_Date')['CSAT Score'].mean()
daily_csat_avg.head()
```

```python
daily_csat_avg.rolling(window=7).mean()
```

```python
daily_csat_avg_rolling = daily_csat_avg.rolling(window=7).mean()
```

```python
# Line plot with rolling average of 7 days
plt.figure(figsize=(12, 6))
sns.lineplot(x=daily_csat_avg.index, y=daily_csat_avg, label='Daily Average CSAT Score')
sns.lineplot(x=daily_csat_avg_rolling.index, y=daily_csat_avg_rolling, label='7-Day Rolling Average CSAT Score')

plt.title('Daily Trend of Average CSAT Score (with Rolling Average)')
plt.xlabel('Date')
plt.ylabel('Average CSAT Score')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Normal line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Survey_response_Date', y='CSAT Score')
plt.title('Daily Trend of Average CSAT Score')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### 23. Plot number of issues reported weekly.
 Bar chart (aggregated)


```python
temporal_df.columns
```

```python
# Dividing into bins
bins = [
    temporal_df['Issue_reported at'].min(),
    '07/08/2023 00:00',
    '14/08/2023 00:00',
    '21/08/2023 00:00',
    '28/08/2023 00:00',
    temporal_df['Issue_reported at'].max()
    ]

labels = ['Week 1', 'Week 2', 'Week 3', 'Week 4', 'Week 5']

temporal_df['Week'] = pd.cut(temporal_df['Issue_reported at'], bins=bins, labels=labels)
temporal_df.head()
```

```python
issues_weekly = temporal_df.groupby('Week')['Week'].count()
issues_weekly
```

```python
# bar chart
sns.barplot(x=issues_weekly.index, y=issues_weekly)
plt.title('Number of Issues Reported Weekly')
plt.show()
```

- Alternative - Histplot

```python
# Hist plot
sns.histplot(data=temporal_df, x='Week')
plt.title('Number of Issues Reported Weekly')
plt.show()
```
